```python
import whisper
import os
import pandas as pd
from tqdm import tqdm

BASE_DIR = "Dataset_tamil/"
DEPRESSED_DIR = os.path.join(BASE_DIR, "Depressed")
NOT_DEPRESSED_DIR = os.path.join(BASE_DIR, "Non-depressed")

OUTPUT_CSV = "tamil_audio_dataset.csv"

# Use medium for safety
model = whisper.load_model("medium", device="cpu")

rows = []

def process_folder(root_folder, label):
    print(f"\n📂 Scanning: {root_folder}")

    for root, _, files in os.walk(root_folder):
        for file in tqdm(files, desc=f"Processing {label}"):
            if file.lower().endswith((".wav", ".mp3", ".m4a", ".flac")):
                audio_path = os.path.join(root, file)
                try:
                    result = model.transcribe(
                        audio_path,
                        language="ta",
                        fp16=False
                    )

                    text = result["text"].strip()
                    print(f"\n🎧 {audio_path}")
                    print("📝", text)

                    rows.append([text, label])

                except Exception as e:
                    print(f"❌ Error: {audio_path} → {e}")

# Run
process_folder(DEPRESSED_DIR, "dep")
process_folder(NOT_DEPRESSED_DIR, "ndep")

print("\n✅ Total samples:", len(rows))

df = pd.DataFrame(rows, columns=["audio_text", "label"])
df.to_csv(OUTPUT_CSV, index=False, encoding="utf-8")

print(f"📄 Saved as {OUTPUT_CSV}")
```
