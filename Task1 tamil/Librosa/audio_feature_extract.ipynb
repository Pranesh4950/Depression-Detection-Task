```python
import os
import librosa
import numpy as np
import pandas as pd
```

```python
def extract_audio_features(audio_path):
    y, sr = librosa.load(audio_path, sr=16000, mono=True)

    # MFCC
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfcc, axis=1)
    mfcc_std = np.std(mfcc, axis=1)

    # Pitch
    pitches, _ = librosa.piptrack(y=y, sr=sr)
    pitch_vals = pitches[pitches > 0]
    pitch_mean = np.mean(pitch_vals) if len(pitch_vals) > 0 else 0
    pitch_std = np.std(pitch_vals) if len(pitch_vals) > 0 else 0

    # Energy
    energy = np.mean(y ** 2)
    energy_std = np.std(y)

    # Silence ratio
    silence_ratio = np.sum(np.abs(y) < 0.02) / len(y)

    features = np.hstack([
        mfcc_mean,
        mfcc_std,
        pitch_mean,
        pitch_std,
        energy,
        energy_std,
        silence_ratio
    ])

    return features
```

```python
dataset_path = "Dataset"

data = []

classes = {
    "Depressed": 1,
    "Non-depressed": 0
}

for class_name, label in classes.items():
    class_folder = os.path.join(dataset_path, class_name, "Train_set")

    for file in os.listdir(class_folder):
        if file.endswith(".wav"):
            file_path = os.path.join(class_folder, file)

            features = extract_audio_features(file_path)

            row = {
                "file_name": file,
                "label": label
            }

            # Add feature columns
            for i, value in enumerate(features):
                row[f"f{i+1}"] = value

            data.append(row)
```

```python
df = pd.DataFrame(data)

print(df.shape)
df.head()
```

```python
df.to_csv("audio_features.csv", index=False)
print("audio_features.csv saved successfully!")
```
